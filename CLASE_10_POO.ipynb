```python
class Casa:
  """
  Construir el constructor
  argumentos:
    color(str): color de la casa
    numero_habitaciones (int): el nmero de habitaciones de la casa
    numero_baos(int): el nmero de baos de la casa
    estacionamiento(bool): si tiene estacionamiento o no
    tiene_patio(bool): si tiene patio o no
  """
  def __init__(self, color, numero_habitaciones, numero_baos, estacionamiento, tiene_patio):
    self.color = color
    self.numero_habitaciones = numero_habitaciones
    self.numero_baos = numero_baos
    self.estacionamiento = estacionamiento
    self.tiene_patio = tiene_patio

  def descripcion(self):
    #Describir la casa
    descripcion=f"La casa es de color {self.color}, tiene {self.numero_habitaciones}, {self.numero_baos} baos "
    if self.estacionamiento:
      descripcion += "y tiene estacionamiento"

    if self.tiene_patio:
      descripcion += " y tiene patio"
    return descripcion

  def pintar(self,nuevo_color):
    self.color=nuevo_color
    print(f"La casa ha sido pintada de color {self.color}")
```

```python
mi_casa=Casa("blanca",4,5,True,True)
type(mi_casa)
```

```python
print(mi_casa.descripcion())
```

```python
mi_casa.pintar("azul")
```
